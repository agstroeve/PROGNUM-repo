```python
import numpy as np 
from pathlib import Path
import os 
from matplotlib.pyplot import figure, subplots, show
from math import sqrt

tilde_path = os.path.expanduser('~') 
subdirectory = 'PROGNUM-repo'
subdirectory2 = 'Task5'
rotcur = 'rotcur.txt'
OSpath = f"{tilde_path}/{subdirectory}/{subdirectory2}/{rotcur}"   #write the path to where the file (rotcur.txt) can be found 
path = Path(OSpath)

mode = 'r'  # reading mode 
r = open(rotcur, mode)  # open the file 


radius, velocity = np.loadtxt(path, comments="#", unpack=True)  # turn the radius and velocity into arrays 
fig = figure()
frame = fig.add_subplot(1,1,1)
frame.plot(radius, velocity)
show()   # plot a graph of radius vs velocity



G = 6.67*10**-11  #gravitational constant 
Msun = 1.99*10**30  # solar mass
M = Msun * 2*10**10   # mass of NGC 6946
new_r = np.delete(radius,0)  # delete the first value of radius as it is 0 (cannot divide by zero)
keplerian = [sqrt((G*M)/ num) for num in new_r]  #formula for the keplerian 

np.savetxt('keplerian.txt', keplerian, fmt='%.18e', delimiter=' ', newline='\n',  #save the keplerian data as a file 
              header='', footer='', comments='# ')  
kep_file = np.loadtxt('keplerian.txt')  # load the keplerian file 


fig = figure()
frame = fig.add_subplot(1,1,1)
frame.plot(kep_file)
show()   # plot a graph of the keplerian data

```

```python
from imageio import imwrite, imread  
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure, show, cm

ngc = np.loadtxt('ngc6946.txt', dtype=str)  # import ngc 
ngc.shape = (40,50,70)  #shape ngc 
image_1 = (ngc[0,:,:])  #take first z-plane 
print(image_1)
image_2 = image_1[:].astype(float)  #change into floats
plt.imshow(image_2, interpolation='none', cmap=cm.bone, origin='lower')  #plot 
plt.colorbar()

mean_i = np.mean(image_2)
std_i = np.std(image_2)
plt.contour(image_2, levels=[mean_i, mean_i+std_i, mean_i+2*std_i])  #draw contour with three levels
show()
```

```python
from scipy.constants import c

class Galaxy:    
    """Class for redshift calculation of galaxies"""  
        
    caH_0 = 3968.5      # Calcium H line in rest frame
    
    def __init__(self, name, calciumHline=caH_0):
        self.name = name
        self.caH = calciumHline
        self.c = c
    def redshift(self):        
        z = (self.caH - self.caH_0)/self.caH
        return z
    def velocity(self):
        velo = self.c*self.redshift()*10**(-3)
        return velo

gal3 = Galaxy("NGC1832", 3994.7)
name = gal3.name
c_ = gal3.c
velo = gal3.velocity()

print(f"Velocity of {name} is {velo} km/h, the speed of light is {c_} m/s")

```

```python
class Galaxy:
    """Class for redshift calculation of galaxies"""
    def __init__(self, name, calciumHline=None):
        self.name = name
        self.caH = calciumHline
    def redshift(self):
        caH_0 = 3968.5  # Calcium H line in rest frame
        z = (self.caH - caH_0)/caH_0
        return z

# Initialize some objects
gal1 = Galaxy("NGC1357")
gal2 = Galaxy("NGC1832", 3994.7)

name = gal2.name
z = gal2.redshift()
print(f"Redshift of {name} is {z}")
```
