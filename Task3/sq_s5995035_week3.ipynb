```python
from math import sqrt
from scipy.constants import G

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

number_of_objects = len(masses)
indxlist = range(number_of_objects)

vel = []

for i in indxlist:
    M = masses[i]
    R = radii[i]
    V_esc = sqrt((2*G*M)/R)
    print(f"For mass {M:.2e} the escape speed is {V_esc}")
    vel.append(V_esc)

object_masses = dict(zip(names, masses))

print(object_masses)

object_masses = { 'Sun': 1.9891e+30, 'Jupiter': 1.8986e+27, 'Saturn': 5.6846e+26, 'Neptune': 1.0243e+26, 'Uranus': 8.6810e+25, 
                 'Earth': 5.9736e+24, 'Venus': 4.8685e+24, 'Mars': 6.4185e+23, 'Mercury': 3.3022e+23, 
                 'Moon': 7.349e+22, 'Pluto': 1.25e22}

planetkeys = object_masses.keys()
for key in planetkeys:
    print(f"{key} has mass {object_masses[key]}")
```

```python
import numpy as np

list_18 = [18.49, 35.48, 35.48, 11.34, 7.12, 9.82, 23.66, 44, 66.7, 52.4, 55, 25.6, 26.19]
array_18 = np.array(list_18)
list_19 = [2094.75, 2320.87, 1224.9, 872.13, 432.2, 751, 1410, 2700, 5292, 2550, 5253, 2419, 2158]
array_19 = np.array(list_19)


sum_x= np.sum(array_18)
x_avg = sum_x/len(array_18)  # average of x-values
sum_y=np.sum(array_19)
y_avg = sum_y/len(array_19)  # average of y-values
print(x_avg)

part_1 = np.sum((array_18 - x_avg)*(array_19 - y_avg))

b = part_1 / np.sum((array_18 - x_avg)**2)
print(f"b is:",b,"")

a = y_avg - b*x_avg
print(f"a is:" ,a,"")
```
