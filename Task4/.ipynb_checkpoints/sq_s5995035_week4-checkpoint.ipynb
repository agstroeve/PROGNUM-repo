```python
import numpy as np
from matplotlib import pyplot as plt
import scipy

lengths = np.random.normal(185, 8, 100000)
print("Subset of random lengths:", lengths[:10])
dev = np.std(lengths)  #standard deviation of lengths 
print(f"Calculated standard deviation:", dev,"")  
mea = np.mean(lengths)  #mean of lengths
print(f"Calculated mean:", mea,"")  

plt.hist(lengths, bins=40, label="$Data\, ")
plt.legend([mea, dev])
x1, y1 = [185, 185], [0, 9000]
plt.plot(x1, y1, marker = 'o')
plt.axvspan(185-8, 185+8, color='yellow', alpha=0.5)
plt.show()

smaller = scipy.stats.norm.cdf(185, loc=185, scale=8) #probability that any height is <= 185cm
print(f"Probability of height smaller than or equal to mean: {smaller}")

ran = scipy.stats.norm.cdf(185+8, loc=185, scale=8) - scipy.stats.norm.cdf(185-8, loc=185, scale=8)
print(f"Probability of height within 1 stdev range: {ran}")
# last 2 questions ? 
```

```python
import numpy as np 

expr = input("Enter function in terms of x: ")  #ask user for integral input
func = lambda x: eval(expr)  #change the input integral to a function so the method recognizes x as variable 
a = float(input("Lower bound:"))  #change the input to floats so they can be used in calculations 
b = float(input("Upper bound:"))

def monte_carlo(func, a, b):   #define function monte carlo, input function, lower bound, and upper bound 
    X = np.random.uniform(a, b, 1000)   #create an array between lower and upper bound, 1000 values 
    Y = np.array([func(i) for i in X])   #y-value at each of the 1000 values    
    return (b-a) * np.mean(Y)  #upper minus lower bound * the average y-value
    
monte_carlo(func, 0, 1)


#download the python script to the pc and in terminal type: python monte_carlo.py
```

```python
import numpy as np
import math 
import matplotlib.pyplot as plt

#rotor = cos(theta) + j*sin(theta)

degs = 60
theta = math.radians(60)  #change 60 degrees to radians 
rotor = math.cos(theta) + 1j*math.sin(theta)  #rotor for angle 60 degrees 
print(rotor)

fig=plt.figure()
ax = fig.add_subplot(1,1,1)
x = np.array([1,3,3,1,1])
y = np.array([1,1,3,3,1])
ax.set_xlim(-2,4)
ax.set_ylim(0,4)
plt.plot(x,y)

z=x+1j*y
z_rot = z*rotor
x_rot = z_rot.real
y_rot = z_rot.imag

plt.plot(x_rot,y_rot)
```

```python
import numpy as np
import math

def funcfilter(func, seq):
    seq = np.asarray(seq)  
    filt = seq[seq>0]  #return all values in seq that are >0
    print(func(filt))  #apply func to the filtered list and print 
    
funcfilter(np.sum, [-1,2,3,4])
funcfilter(np.log10, np.array([-100, -10, 0, 10, 100]))
funcfilter(np.log10, 100)
```

```python
from astropy.cosmology import Planck18
import numpy as np 
import scipy
from math import sqrt

H0 = Planck18.H0.value
Om0 = Planck18.Om0
Ode0 = Planck18.Ode0


def Hz(z):
    F = H0*sqrt(Om0*(1+z)**3+Ode0)
    return F

def tLz(z):
    G = 1/((1+z)*Hz(z))
    return G

z=2
tL, err = scipy.integrate.quad(tLz, 0, z)
print(f"{tL} km / (Mpc/s)")  
tL_sec = tL * 3.09*(10**19)
tL_Gyr = tL_sec * 3.1710 * (10**(-17))
print(f"{tL_Gyr} Gyr")
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import gaussian_kde

X = np.random.normal(0, 1, 100000)
Y = np.random.normal(2, 1, 100000)

kde_x = gaussian_kde(X)  #calculate Kernel Density Estimate of x 
kde_y = gaussian_kde(Y)

x = np.linspace(min(X)-1, max(X)+1, 1000)  #create range of values where to evaluate kde 
y = np.linspace(min(Y)-1, max(Y)+1, 1000)
pdf_x=kde_x(x)  #evaluate kde at x 
pdf_y=kde_y(y)


fig=plt.figure()
ax = fig.add_subplot(1,1,1)
plt.hist(X, bins=30, alpha=0.5, color='blue', edgecolor='black', density=True, label='X')  
plt.hist(Y, bins=30, alpha=0.5, color='red', edgecolor='black', density=True, label='Y')
plt.plot(x, pdf_x, color='black', linestyle='dotted', linewidth=6, label='X(PDF)')
plt.plot(y, pdf_y, color='black', linestyle='dashed', linewidth=6, label='Y(PDF)')
plt.ylabel('Probability Density')
plt.xlabel('Values')
plt.legend()
plt.annotate(
    'N(0,1)',               # text
    xy=(-2, 0.15),           # arrow points here (x, y)
    xytext=(-3, 0.15),       # location of the text
    arrowprops=dict(facecolor='blue', arrowstyle='-|>', lw=3),
    color='blue',
    fontsize=12,
    ha='right')
plt.annotate(
    'N(2,1)',
    xy=(4, 0.15),
    xytext=(6.5, 0.15),
    arrowprops=dict(facecolor='blue', arrowstyle='-|>', lw=3),
    color='red',
    fontsize=12,
    ha='right')
plt.title('1D Histogram + PDF')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import gaussian_kde

X = np.random.normal(0, 1, 100000)
Y = np.random.normal(2, 1, 100000)

xy = np.vstack([X, Y])  #stack the two arrays 
kde = gaussian_kde(xy)  #evaluate the kde 

xgrid = np.linspace(min(X), max(X), 50)  #create a grid
ygrid = np.linspace(min(Y), max(Y), 50) 
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)  #combine to create a 2d grid 

Z = kde(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))  #ravel to flatten the arrays, then stack them
Z = Z.reshape(Xgrid.shape)  #reshape stacked array into grid 

plt.figure()
plt.hist2d(X, Y, bins=30, cmap='Blues', density=False)  
plt.colorbar()

levels = np.max(Z) * np.array([
    np.exp(-0.5 * 3**2),   # 3σ  #create three contour for the different standard deviation levels 
    np.exp(-0.5 * 2**2),  # 2σ
    np.exp(-0.5 * 1**2)  # 1σ
])

plt.contour(
    Xgrid, Ygrid, Z,  #plot the grid and values
    levels=levels,  #contour drawn at previously defined levels 
    colors=['red'],  #all lines red 
    linewidths=2
)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Histogram + Contour')

plt.grid(alpha=0.3)
plt.show()
```
