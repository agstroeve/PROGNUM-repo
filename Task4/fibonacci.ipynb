```python
import numpy as np
N=100
seq= [0,1]
for i in range(2,N):
    next = seq[-1] + seq[-2]
    seq.append(next)  #fibonacci sequence
s=sum(seq[1::2])  #adding all numbers wth odd index
for i in range (2,int(np.sqrt(float(s)))):  #is it a prime numbre
    if s%1 == 0:  
        print(f"not prime")
        break
```
