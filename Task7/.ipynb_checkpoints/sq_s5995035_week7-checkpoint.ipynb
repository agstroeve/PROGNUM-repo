```python
import numpy as np 
from numpy import linalg as la 

A1 = np.array((1, 2, 3, 0, -4, 1, 0, 3, -1)).reshape(3,3)  #create vector A1 
print(A1)
print() 
 
det_A1 = la.det(A1)  #determinant of A1 
print(f"Determinant = {det_A1}")
print()

# A.x = B so x = inv(A).B
B1= np.array((2, 2, -1))  #create a 1d vector which represents the right side of the equation
xyz = la.inv(A1)@B1  #dot product Ainv and B1 
print(f"X, Y, Z =", xyz,"")   
print()

# A1 is orthogonal if A.Atranspose = I and inv(A) = Atranspose
A1_t = A1.T  #transpose of A1 
print(f"Transpose of A1 =\n", A1_t,"")
print(f"Inverse of A1 =\n", la.inv(A1),"")
print(f"The transpose and inverse of A1 are not equal so A1 is not orthogonal") 
print()

Lambda1, vA = la.eig(A1)  #Calculate eigenvalues and eigenvectors 
print("Eigenvalues = ", Lambda1)
print(f"\nEigenvectors:\n{vA}")
print()
vA1, vA2, vA3 = vA.T  #unpack eigenvector into three vectors
print(f"Individual eigenvectors:") 
print(vA1)
print(vA2)
print(vA3)
```

```python
from math import cos, sin, radians

def Ex(ang):
    rad_A = radians(ang)
    Ex_rot = np.array([1, 0, 0, 0, cos(rad_A), sin(rad_A), 0, -sin(rad_A), cos(rad_A)]).reshape(3,3)
    return Ex_rot

def Ey(ang):
    rad_A = radians(ang)
    Ey_rot = np.array([cos(rad_A), 0, -sin(rad_A), 0, 1, 0, sin(rad_A), 0, cos(rad_A)]).reshape(3,3)
    return Ey_rot

def Ez(ang):
    rad_A = radians(ang)
    Ez_rot = np.array([cos(rad_A), sin(rad_A), 0, -sin(rad_A), cos(rad_A), 0, 0, 0, 1]).reshape(3,3)
    return Ez_rot

print(f"\nEx(33)=\n{Ex(33)}")
print(f"\nEy(33)=\n{Ey(33)}")
print(f"\nEz(33)=\n{Ez(33)}")
```

```python
import numpy as np

def bilinearIP(I, col, row): # Make it X, Y compatible
    rowlen, collen= I.shape
    i = int(row)
    j = int(col)
    if row - i == 0 and col - j == 0:  # Falls exactly on a gridpoint
        return I[i,j]

    inbounds = (0 <= j <= collen-1) and (0 <= i <= rowlen-1)
    if not inbounds:
        Iintp = None
    else:
        mu =  row - i
        Lambda = col - j
        # If we are at the last pixel at the right or bottom, then there
        # is not a neighbour, so we take the pixel itself as the neighbour
        I0 = I[i,j]
        I1 = I[i,min(j+1,collen-1)]
        I2 = I[min(i+1,rowlen-1),j]
        I3 = I[min(i+1,rowlen-1),min(j+1,collen-1)]
        Iintp = (1-Lambda)*(1-mu)*I0 + Lambda*(1-mu)*I1 + mu*(1-Lambda)*I2 + Lambda*mu*I3
    return(Iintp)

I = np.array([[2.0,4.1], [2.5,3.8]])

row_y = 0.7 
col_x = 0.2  

Iintp = bilinearIP(I, col_x, row_y)
if Iintp:
    print(f"The interpolated intensity at i,j={row_y},{col_x} => x, y={col_x}{row_y} is {Iintp:.6f}")
else:
    print("One of the indices was outside the limits")
```

```python
import numpy as np
from scipy.optimize import leastsq, curve_fit
import matplotlib.pyplot as plt 

data = [
   18.49,     2094.75,    0.3,           100,
   35.48,     2320.87,    2.2,           120,
   35.48,     1224.9,     3.0,           80,
   11.34,     872.13,     1.2,           80,
   7.12,      432.2,      0.4,           70,
   9.82,      751,        0.3,           75,
   23.66,     1410,       2.3,           70,
   44,        2700,       4.0,           110,
   66.7,      5292,       5.3,           120,
   52.4,      2550,       5.2,           100,
   55,        5253,       6.1,           150,
   25.6,      2419,       3.2,           110,
   26.19,     2158,       2.8,           110
]

data = np.asarray(data)
d = data[0::4]; v = data[1::4]  #separate data columns 
err_d= data[2::4]; err_v= data[3::4]

beta0=0  #beta0 is 0 for hubble's law 

def model(beta, d):  #model function 
    beta0, beta1 = beta
    return beta0 + beta1*d

def residuals(beta, d, v, err_v):
    return (v - model(beta,d))/err_v

def model_cf(x, beta0, beta1):  #model function for curve fit 
    """The model with parameters a and b. Note that for the Hubble relation we fix a to 0.0"""
    return beta0 + beta1*x

beta_c, pcov = curve_fit(model_cf, d, v, sigma= err_v, absolute_sigma=True)  #curve fit 
print(f"Intercept beta0={beta_c[0]}, slope beta1={beta_c[1]}")

X = np.array([d.min(),d.max()])  #calculate end points for curvefit
Y = model_cf(X, beta_c[0], beta_c[1])



beta_start = [0, 70.0]         # initial estimate 
data = (d, v, err_v)             # pack data into tuple 
result = leastsq(residuals, beta_start, args=data, full_output=1)  #apply leastsquares 
(beta, pcov, infodict, errmsg, ier) = result       # unpack results 
print(result)  #print results 

fig = plt.figure()
frame = fig.add_subplot(1,1,1)
plt.scatter(d, v, marker='.')  #plot given data 
frame.grid(True)

dmin = d.min(); dmax = d.max()  #set min and max values from data 
vmin = model(beta,dmin); vmax = model(beta,dmax)  #apply model 
frame.plot([dmin, dmax], [vmin, vmax], label=f"leastsq: v = {beta[0]:.2f} + {beta[1]:.2f} d", color='orange')
frame.plot(X, Y, label=f"curve: v = {beta_c[0]:.2f} + {beta_c[1]:.2f} d", color='green')
frame.legend()
plt.show()
```
