```python
from math import atan2, degrees, tan
X = eval(input("Input a value for X:" ))
Y = eval(input("Input a value for Y:" ))
angle = atan2(Y, X)  #angle in radians
angle_d = degrees(angle)
print(f"The angle in degrees is:", angle_d,)
if 0 < angle_d < 90:
    print(f"{X,Y} belongs to quadrant 1")
elif 90 < angle_d < 180:
    print(f"{X,Y} belongs to quadrant 2")
elif -180 < angle_d < -90:
    print(f"{X,Y} belongs to quadrant 3")
elif -90 < angle_d < 0:
    print(f"{X,Y} belongs to quadrant 4") 
Z = tan(angle)
print(f"The tangent of the angle is", Z,)
```

```python
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]
night1s = set(night1)
night2s = set(night2)
inters = set.intersection(night1s, night2s)
inters_list = list(inters)
print(f"Intersection of night 1 and night 2: {inters_list}")
if 5 in inters_list:
    print(f"5 is part of the intersection list") 
else:
    print(f"5 is not part of the intersection list")
print(night1s)
if 5 in night1s:
    print(f"5 is a part of the set version of night 1")
else:
    print(f"5 is not a part of the set version of night 1") 
```
